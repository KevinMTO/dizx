```python
zcircuit = dizx.Circuit(3)
zcircuit.add_gate("Z", 0)
zcircuit.add_gate("H", 1)
zcircuit.add_gate("NEG", 0)
zcircuit.add_gate("CX", 1, 2)
zcircuit.add_gate("CZ", 0, 1)
zcircuit.add_gate("NEG", 1)
zcircuit.add_gate("SWAP", 0, 1)
from dizx.circuit.gates import S
zcircuit.add_gate(S(2,adjoint=True))
g = zcircuit.to_graph(zh=True)
dizx.draw(g)
```

```python
from dizx.simplify import to_gh

to_gh(g)
dizx.draw(g)
```
