```python
zcircuit = dizx.Circuit(3)
zcircuit.add_gate("Z", 0)
zcircuit.add_gate("H", 1)
zcircuit.add_gate("NEG", 0)
zcircuit.add_gate("S", 1)
zcircuit.add_gate("CX", 1, 2)
zcircuit.add_gate("CZ", 0, 1)
zcircuit.add_gate("CZ", 0, 1)
zcircuit.add_gate("NEG", 1)
from dizx.circuit.gates import S
zcircuit.add_gate(S(2,adjoint=True))
g = zcircuit.to_graph(zh=True)
dizx.draw(g)
```

```python
from dizx.simplify import to_graph_like

to_graph_like(g)
dizx.draw(g, labels=True)
```

```python
from dizx.rules import local_complementation_simplification

local_complementation_simplification(g, 11)
dizx.draw(g, labels=True)
```

```python
zcircuit = dizx.Circuit(2)
zcircuit.add_gate("NEG", 1)
zcircuit.add_gate("S", 1)
zcircuit.add_gate("S", 1)
zcircuit.add_gate("CZ", 0, 1)
zcircuit.add_gate("CZ", 0, 1)
zcircuit.add_gate("CX", 0, 1)
g = zcircuit.to_graph(zh=True)
to_graph_like(g)
dizx.draw(g, labels=True)
```

```python
from dizx.rules import local_complementation_simplification

local_complementation_simplification(g, 5)
dizx.draw(g, labels=True)
```

```python
# EDGES:
# z = 1  =>  z^(-1) = 1
# -z^(-1) e_k e_j = -e_k e_j, that's easy check
# PHASES:
# a = 0  =>  x=0
# y = y - e^2:
# 0 - 1 = 1 GOOD
# 1 - 1 = 0 GOOD
# 1 - 2^2 = 1 - 2 = 0 GOOD

local_complementation_simplification(g, 2)
dizx.draw(g, labels=True)
```
