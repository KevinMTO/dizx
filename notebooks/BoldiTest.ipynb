```python
import dizx
from dizx import Edge
Phase = dizx.CliffordPhase
d = 3
Z = dizx.VertexType.Z
X = dizx.VertexType.X
```

```python
g = dizx.Graph(d)
```

```python
v = g.add_vertex(Z,qubit=1,row=0)
phase = Phase(d,1)
w = g.add_vertex(X,qubit=2,row=0,phase=phase)
g.add_edge(g.edge(v,w),Edge(had=2))
x = g.add_vertex(X,qubit=2,row=2,phase=phase)
g.add_edge(g.edge(w,x),Edge(had=2))
```

```python
z = g.add_vertex(0,0,0)
g.add_edge(g.edge(z,v),eo:=dizx.Edge(simple=1))
```

```python
dizx.draw(g)
```

```python
from dizx.basicrules import x_color_change

x_color_change(g, w)
dizx.draw(g)
```

```python
x_color_change(g, x)
dizx.draw(g)
```

```python
zcircuit = dizx.Circuit(3)
zcircuit.add_gate("Z", 0)
zcircuit.add_gate("H", 1)
zcircuit.add_gate("NEG", 0)
zcircuit.add_gate("CX", 1, 2)
zcircuit.add_gate("CZ", 0, 1)
zcircuit.add_gate("SWAP", 0, 1)
from dizx.circuit.gates import S
zcircuit.add_gate(S(2,adjoint=True))
g = zcircuit.to_graph()
dizx.draw(g)
```
